```c++
#include "JupyterOpencvIncludeLibraries.h"
#include "displayImages.h"
#include <opencv2/opencv.hpp>
using namespace cv;
```

```c++
int inputH = 100;
int inputW = 100;
cv::Mat demoImageC1(inputH, inputW, CV_8UC1, cv::Scalar(0, 0));
cv::Mat demoImageC3(inputH, inputW, CV_8UC3, cv::Scalar(0, 0,0));

cv::circle(demoImageC1, cv::Point(20, 20), 10, 255, -1);
cv::rectangle(demoImageC1,cv::Point(40, 40),cv::Point(60, 60),255,-1);
cv::circle(demoImageC1, cv::Point(80, 80), 15, 255, -1);
displayImage(demoImageC1);

cv::circle(demoImageC3, cv::Point(20, 20), 10, cv::Scalar(255,0,0), -1);
cv::rectangle(demoImageC3,cv::Point(40, 40),cv::Point(60, 60),cv::Scalar(0,255,0),-1);
cv::circle(demoImageC3, cv::Point(80, 80), 15, cv::Scalar(0,0,255), -1);
displayImage(demoImageC3);
```

## 利用at访问Mat中的元素

该方法需要提前预知Mat中的元素类型，否则读取出错

### 单通道

```c++
std::cout<<demoImageC1.rows<<","<<demoImageC1.cols<<std::endl;
for( size_t nrow = 0; nrow < demoImageC1.rows; nrow++)  
{  
    for(size_t ncol = 0; ncol < demoImageC1.cols; ncol++)  
    {  
        uchar val = demoImageC1.at<uchar>(nrow,ncol);
        std::cout<<(int)val<<",";
    } 
    std::cout << std::endl;
}  
```

### 三通道

```c++
for( size_t nrow = 0; nrow < demoImageC3.rows; nrow++)  
{  
    for(size_t ncol = 0; ncol < demoImageC3.cols; ncol++)  
    {  
        Vec3i bgr = demoImageC3.at<Vec3b>(nrow,ncol);//用Vec3b也行  
        std::cout   << "("<<bgr.val[0]<<","  <<bgr.val[1]<<","  <<bgr.val[2]<<")";  
    }  
    std::cout << std::endl;  
}  
```

## 使用指针读取

```c++
for(size_t nrow=0;nrow<demoImageC3.rows;nrow++){
    uchar* data=demoImageC3.ptr<uchar>(nrow);
    for(size_t ncol=0;ncol<demoImageC3.cols*demoImageC3.channels();ncol++){
       std::cout<<int(data[ncol])<<",";
    }
   std::cout<<std::endl;
}
```

## 使用迭代器

```c++
MatIterator_<uchar> it=demoImageC3.begin<uchar>();
MatIterator_<uchar> it_end=demoImageC3.end<uchar>();
for(int cnt=1;it!=it_end;++it){
    std::cout<<(int(*it))<<",";
    if((cnt++%demoImageC3.cols)==0){
        std::cout<<std::endl;
    }
}
```

## 使用矩阵地址

```c++
for(size_t nrow=0;nrow<demoImageC3.rows;nrow++){
    for(size_t ncol=0;ncol<demoImageC3.cols;ncol++){
        std::cout<<(int)(*(demoImageC3.data+demoImageC3.step[0]*nrow+demoImageC3.step[1]*ncol))<<",";
    }
    std::cout<<std::endl;
}
```
